```python
from matplotlib import pyplot as plt
from matplotlib import cm
from numpy import matrix
import gensim
from urllib.request import urlopen
import pickle
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
count = CountVectorizer()
docs = np.array
import pyLDAvis.gensim
```

# Tf-idf by gensim in corpus_5

## download data

```python
corpus5=pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-03/raw/master/Data/main/corpus_5.pkl'))
dictionary5=pickle.load(urlopen('https://github.com/Galeforse/DST-Assessment-03/raw/master/Data/main/dictionary_5.pkl'))
```

## Tf-idf with different parameters

```python
tfidf5 = gensim.models.TfidfModel(corpus5)
corpus_tfidf5 = tfidf5[corpus5]
```

### 10topics and 1 iteration

```python
model_tfidf51 = gensim.models.LdaMulticore(corpus_tfidf5, num_topics=10, id2word=dictionary5, passes=2, workers=4)
model_tfidf51.show_topics(10,7)
```

### 10topics and 50 iterations

```python
model_tfidf52 = gensim.models.LdaMulticore(corpus_tfidf5, num_topics=10, id2word=dictionary5, passes=2, workers=4,iterations=50)
model_tfidf52.show_topics(10,7)
```

### 5topics and 1 iteration

```python
model_tfidf53 = gensim.models.LdaMulticore(corpus_tfidf5, num_topics=5, id2word=dictionary5, passes=2, workers=4)
model_tfidf53.show_topics(5,7)
```

### 5topics and 50 iteration

```python
model_tfidf54 = gensim.models.LdaMulticore(corpus_tfidf5, num_topics=5, id2word=dictionary5, passes=2, workers=4,iterations=50)
model_tfidf54.show_topics(5,7)
```

## perplexities and cohenrence

```python
log_perplexities = {
        "model_tfidf51" : model_tfidf51.log_perplexity(corpus_tfidf5[0:1000]),
     "model_tfidf52" : model_tfidf52.log_perplexity(corpus_tfidf5[0:1000]),
     "model_tfidf53" : model_tfidf53.log_perplexity(corpus_tfidf5[0:1000]),
     "model_tfidf54" : model_tfidf54.log_perplexity(corpus_tfidf5[0:1000])
    };
log_perplexities
```

```python
from gensim.models.coherencemodel import CoherenceModel
def getCoherence(m,c,d):
    coherence_model_lda = CoherenceModel(model=m,corpus=c, dictionary=d, coherence='u_mass')
    coherence_lda = coherence_model_lda.get_coherence()
    return(coherence_lda)
```

```python
coherences={
    "model_tfidf51": getCoherence(model_tfidf51,corpus_tfidf5[0:1000],dictionary5),
    "model_tfidf52": getCoherence(model_tfidf52,corpus_tfidf5[0:1000],dictionary5),
    "model_tfidf53": getCoherence(model_tfidf53,corpus_tfidf5[0:1000],dictionary5),
    "model_tfidf54": getCoherence(model_tfidf54,corpus_tfidf5[0:1000],dictionary5)
}
coherences
```
